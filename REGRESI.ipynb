```python
import pandas as pd
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

pandas dan numpy untuk manipulasi data.

matplotlib.pyplot untuk visualisasi grafik.

sklearn.linear_model.LinearRegression untuk membuat model regresi linier.

```python
dataset = pd.read_csv('R04_corruptions.csv')
dataset
```

Penjelasan:

Dataset dibaca dari file CSV bernama R04_corruptions.csv dan disimpan dalam variabel dataset



```python
dataset.columns = ['x', 'y']
dataset
```

Penjelasan:

Nama kolom dalam dataset diubah menjadi x dan y untuk memudahkan akses.


```python
dataset.corr()
```

Penjelasan:

Menghitung dan menampilkan matriks korelasi antara kolom-kolom dalam dataset


```python
x_train = dataset['x'].values[:, np.newaxis]
y_train = dataset['y'].values
```

Penjelasan:
Data dibagi menjadi variabel independen (x_train) dan dependen (y_train). x_train diproses dengan np.newaxis untuk mengubahnya menjadi array dua dimensi, yang diperlukan oleh LinearRegression..


```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```

Penjelasan:

Sebuah objek model regresi linear dibuat dan dilatih menggunakan data x_train dan y_train


```python
x_test = [[0.5], [1.0], [1.5], [2.0], [2.5], [3.0]]  
predictions = lm.predict(x_test)

print('Coefficient :', lm.coef_)
print('Intercept :', lm.intercept_)
```

Penjelasan:
Model diuji dengan beberapa nilai x_test untuk menghasilkan prediksi. predictions berisi hasil prediksi nilai y untuk setiap x_test.


```python
y_pred_train = lm.predict(x_train)
dfc = pd.DataFrame({'x': dataset['x'], 'y': y_pred_train})
```

Penjelasan:
y_pred_train adalah prediksi nilai y berdasarkan x_train. dfc adalah DataFrame baru yang menyimpan data x dan nilai y hasil prediksi untuk memudahkan proses plotting.


```python
plt.scatter(dataset['x'], dataset['y'], label="Data asli")
plt.plot(dataset['x'], dfc['y'], color='red', linewidth=1, label="Regression Line")
plt.xlabel('Perception')
plt.ylabel('Transparency')
plt.legend()
plt.show()
```

Penjelasan:
Data asli diplot sebagai titik-titik scatter, sedangkan garis regresi ditambahkan dengan plt.plot. Hasilnya, grafik menunjukkan hubungan antara x dan y dengan garis regresi sebagai prediksi model.


```python
y_asli = [0.3, 1.5, 0.9]
y_pred_hasil = lm.predict([[0.45], [0.56], [0.78]])
print(y_pred_hasil)
```

Penjelasan:
y_pred_hasil adalah hasil prediksi untuk beberapa nilai x baru. Selanjutnya, nilai asli y_asli dibandingkan dengan y_pred_hasil untuk menghitung akurasi model.
Melakukan prediksi untuk nilai 0.45, 0.56, dan 0.78, dan mencetak hasil prediksi.


```python
from sklearn.metrics import r2_score
akurasi = r2_score(y_asli, y_pred_hasil)
print(akurasi*100)
```

Penjelasan:

r2_score menghitung koefisien determinasi antara nilai prediksi dan nilai asli, lalu dikalikan dengan 100 untuk menampilkan hasil dalam persentase. Nilai ùëÖ2 menunjukkan seberapa baik model memprediksi data ‚Äì semakin dekat ke 100%, semakin baik.
